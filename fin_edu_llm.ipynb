```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, ServiceContext
from llama_index.llms.huggingface import HuggingFaceLLM
from llama_index.core import PromptTemplate
```

```python
system_prompt = """
You are a financial education Q&A assistant named Finedu.ai.
Your primary goal is to help beginners understand financial terminologies by providing clear, concise, and accurate explanations.
When defining terms, ensure that your explanations are easy to understand, avoiding complex jargon.
Where applicable, include examples to illustrate the concepts. If a term has multiple interpretations, briefly mention the most common ones.
Always maintain a friendly and supportive tone to encourage learning.
"""

query_wrapper_prompt = PromptTemplate(
    "<|SYSTEM|>{system_prompt}<|USER|>{query_str}<|ASSISTANT|>{context_str}"
)
```

```python
query_wrapper_prompt = query_wrapper_prompt.format(

    system_prompt = system_prompt,
    query_str = "{query_str}",
    context_str = "{context_str}"
)
```

```python
!huggingface-cli login
```

```python
llm = HuggingFaceLLM(
    context_window=4096,
    max_new_tokens=256,
    generate_kwargs={"temperature": 0.0, "do_sample": False},
    system_prompt=system_prompt,
    query_wrapper_prompt=query_wrapper_prompt,
    tokenizer_name="meta-llama/Llama-2-7b-chat-hf",
    model_name="meta-llama/Llama-2-7b-chat-hf",
    device_map="auto",
    model_kwargs={
        "torch_dtype": torch.float16,
        "load_in_8bit": True
    }
)
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from llama_index.core import Settings
from llama_index.embeddings.langchain import LangchainEmbedding

embed_model = LangchainEmbedding(
    HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")
)

Settings.embed_model = embed_model
```

```python
from llama_index.core import Settings

Settings.chunk_size = 1024
Settings.llm = llm
Settings.embed_model = embed_model
```

```python
from llama_index.core import Document

documents = []
with open('/content/financialdata.jsonl', 'r') as file:
    for line in file:
        data = json.loads(line)

        text = f"""
Term: {data.get('term', 'N/A')}
Definition: {data.get('definition', 'N/A')}
Significance: {data.get('significance', 'N/A')}
Usage: {data.get('usage', 'N/A')}
"""
        if 'implications' in data and isinstance(data['implications'], dict):
            text += "Implications:\n"
            text += f"- Increasing: {data['implications'].get('increasing', 'N/A')}\n"
            text += f"- Decreasing: {data['implications'].get('decreasing', 'N/A')}"

        doc = Document(
            text=text,
            metadata={
                "term": data.get("term", "N/A"),
                "usage": data.get("usage", "N/A")
            }
        )
        documents.append(doc)
```

```python
documents
```

```python
index = VectorStoreIndex.from_documents(

    documents,
    llm = llm,
    embed_model = embed_model,
    chunk_size = 1024
)
```

```python
index
```

```python
query_engine = index.as_query_engine()
```

```python
response = query_engine.query("what is eps?")
```

```python
import gradio as gr

def process_query(user_input):

    response = query_engine.query(user_input)
    return response

interface = gr.Interface(
    fn=process_query,
    inputs="text",
    outputs="text",
    title="FinEdu.AI",
    description="Your friendly guide to mastering financial terms with ease. \n Finedu.ai is an interactive chatbot designed to demystify financial terminology for beginners. It offers clear, concise explanations and practical examples, empowering users to confidently navigate the financial landscape. By providing accessible insights into complex concepts, Finedu.ai serves as a valuable educational tool for those seeking to enhance their financial literacy."
)

interface.launch()
```

```python
import gradio as gr

def process_query(user_input):
    response = query_engine.query(user_input)
    return response

custom_theme = gr.themes.Base(
    primary_hue=gr.themes.colors.cyan,
    secondary_hue=gr.themes.colors.indigo,
    neutral_hue=gr.themes.colors.gray,
    spacing_size="lg",
    radius_size="lg",
    text_size="lg"
)

with gr.Blocks(theme=custom_theme) as demo:
    with gr.Row():
        gr.HTML(
            """
            <h1 style='text-align: center; margin-bottom: 1rem; font-size: 3.5rem; font-weight: 700; color: #00B4D8;'>
            FinEdu.AI
            </h1>
            """
        )

    with gr.Row():
        gr.HTML(
            """
            <p style='text-align: center; font-size: 1.2rem; color: #888888; margin-bottom: 2rem;'>
            Your friendly guide to mastering financial terms with ease
            </p>
            """
        )

    with gr.Row():
        with gr.Column(scale=2):
            input_text = gr.Textbox(
                label="Ask your question",
                placeholder="Example: What is Earnings Per Share?",
                lines=3
            )
            submit_btn = gr.Button("Get Answer", variant="primary")

        with gr.Column(scale=3):
            output_text = gr.Textbox(label="Answer")

    gr.Examples(
        examples=[
            "What is Earnings Per Share?",
            "Explain Price-to-Earnings Ratio",
            "Define Market Capitalization"
        ],
        inputs=input_text
    )

    submit_btn.click(
        fn=process_query,
        inputs=input_text,
        outputs=output_text
    )

demo.launch(share=True)
```
